```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split ,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
credit_card =pd.read_csv('creditcard.csv')
credit_card.head()
```

About the values is there is a null values or not


```python
credit_card.info()
```

```python
credit_card['Class'].value_counts()
```

Here 0--->normal Transaction and 1--->fraud Transcation

```python
legal=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
```

```python
print(legal.shape)
print(fraud.shape)
```

```python
legal.Amount.describe()
```

```python
credit_card.groupby('Class').mean()
```

UNDER-SAMPLING


```python
legal_sample=legal.sample(n=492)
```

CONCATING THE TWO EQUAL LEGAL AND FRAUD DATA SET AS LEGAL IS TAKEN RANDOM FORM 28K ROWS


```python
new_dataset=pd.concat([legal_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

MODEL TRAINING LOGISTIC REGRESSION MODEL

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

EVALUTAION

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python

print("ther Acurracy of LogisticRegression:",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("The Acurracy of TEST LogisticRegression:", test_data_accuracy)
```

```python
model=RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("ther Acurracy of RadomForest:",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("The Acurracy of TEST RandomForest:", test_data_accuracy)
```

```python
rf = RandomForestClassifier(random_state=42)

# Define hyperparameter grid for tuning
param_grid = {
    'n_estimators': [100, 200],            # number of trees
    'max_depth': [10, 20, None],           # max depth of each tree
    'min_samples_split': [2, 5, 10],       # min samples required to split
    'min_samples_leaf': [1, 2, 4]          # min samples required at each leaf
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, Y_train)

# Get best model from grid search
best_rf = grid_search.best_estimator_

# Train best model on full training data
best_rf.fit(X_train, Y_train)

# Predictions and accuracy
train_accuracy = accuracy_score(Y_train, best_rf.predict(X_train))
test_accuracy = accuracy_score(Y_test, best_rf.predict(X_test))

print(f"The Accuracy of the best RandomForest on training set: {train_accuracy}")
print(f"The Accuracy of the best RandomForest on test set: {test_accuracy}")
print("Best Hyperparameters:", grid_search.best_params_)
```
